```python
# https://learning.quantum-computing.ibm.com/course/basics-of-quantum-information/single-systems#states-measurements-and-operations
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit
from numpy import sqrt
```

```python
u = Statevector([1/sqrt(2), 1/sqrt(2)])
v = Statevector([(1 + 2.0j) / 3, -2/3])
w = Statevector([1/3, 2/3])
```

```python
display(u.draw("latex"))
display(v.draw("latex"))
display(w.draw("text"))
```

```python
display(u.is_valid())
display(w.is_valid())
v.measure()
```

```python
statistics = v.sample_counts(1000)
display(statistics)
plot_histogram(statistics)
```

```python
X = Operator([
  [0, 1],
  [1, 0]
])

Y = Operator([
  [0, -1j],
  [1j, 0]
])

Z = Operator([
  [1, 0],
  [0, -1]
])

H = Operator([
  [1/sqrt(2), 1/sqrt(2)],
  [1/sqrt(2), -1/sqrt(2)]
])

S = Operator([
  [1, 0],
  [0, 1j]
])

T = Operator([
  [1, 0],
  [0, (1 + 1j) / sqrt(2)]
])

v = Statevector([1, 0])

display(v.draw("latex"))
v = v.evolve(H)
display(v.draw("latex"))
```

```python
circuit = QuantumCircuit(1)

circuit.h(0)
circuit.t(0)
circuit.h(0)
circuit.t(0)
circuit.z(0)

circuit.draw("mpl")
```

```python
ket0 = Statevector([1, 0])
v = ket0.evolve(circuit)
v.draw("latex")
```

```python
statistics = v.sample_counts(4000)
plot_histogram(statistics)
```
